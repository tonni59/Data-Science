```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [34,35,36,38,37,34,33]
avg_t = [45,48,47,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temparature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

#plt.legend(loc="upper right")
#plt.legend(loc="best",shadow=True,fontsize="large")
plt.legend(loc="best",shadow=True,fontsize="small")

plt.grid()
```

**Bar Chart**

```python
import numpy as np
```

```python
company = ["HFGD","JH","IUY","EIRU"]
revenue = [90,136,50,27]
profit = [40,8,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)

plt.xlabel("Company Name")
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")

plt.bar(xpos-0.2,revenue,width = 0.4 ,label="Revenue")
plt.bar(xpos+0.2,profit,width = 0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(xpos,company)

plt.ylabel("Company Name")
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")

plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

**Histogram**

```python
blood_sugar = [115,50,122,35,118,145,55,67,89,170,180]

plt.hist(blood_sugar,bins = 10, rwidth=0.95)
```

```python
blood_sugar_men = [115,50,122,35,118,145,55,67,89,170,180]
blood_sugar_women = [33,56,78,90,120,140,125,178,67,78,90,105,165,190]

plt.xlabel("sugar range")
plt.ylabel("no. of paitent")
plt.title("Blood sugar analysis")

plt.hist([blood_sugar_men,blood_sugar_women],bins = [80,100,125,200], rwidth=0.95,color=['green','orange'],label=['men','women'])

plt.legend()
```

**Pie Chart**

```python
exp_vals = [1400,500,409,600,250]
exp_labels = ["Home rent","Food","Phone?Internet skill","Car","Other utilities"]
```

```python
plt.axis("equal")

plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.1f%%",shadow=True)
plt.show()
```

```python
plt.axis("equal")

plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.1f%%",shadow=True,explode=[0,0.5,0,0.2,0])
plt.show()
```

```python
plt.axis("equal")

plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.1f%%",shadow=True,explode=[0,0.5,0,0.2,0],startangle=45)
plt.show()
```

```python
plt.axis("equal")

plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%0.1f%%",shadow=True,explode=[0,0.5,0,0.2,0],startangle=45)
plt.show()

plt.savefig("piechart.png",bbox_inches="tight",pad_inches=True,transparent=True)
plt.savefig("F:\piechart.pdf",bbox_inches="tight",pad_inches=True,transparent=True)
```
